```python
# Import dependencies.
import matplotlib.pyplot as plt
import pandas as pd
import statistics 
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create line graph plot
plt.plot(x_axis, y_axis)
```

```python
# Create the line graph with ax.plt(), object oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the line graph with ax.plt() ,object oriented interface, a second way
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the line graph with ax.plt() ,object oriented interface, a third way
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot. 
#plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create the line with annotations and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create line graph with annotaions and styling
plt.plot(x_axis, y_axis, marker="p", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create bar chart
plt.bar(x_axis, y_axis)
```

```python
# Create bar chart with annotations and styling
plt.bar(x_axis, y_axis, color="blue", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the horizontal bar chart / stacked bar chart
plt.barh(x_axis, y_axis)
```

```python
# Create a horizonal bar chart with the axis flipped
plt.barh(y_axis, x_axis)
# this chart does not make sense for this data, incuded to demonstrate how do it in python
```

```python
# Create a horizontal bar chart with months in reverse order
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
#invert / switch order of months
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fare ($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Create bar chart with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#Create a regular scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
#create a bubble scatter plot
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#make the bubbles larger 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
#run new bubble scatter plot with largers bubbbles
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#simpler way to accomplish the same task above
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#scatter plots with object oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#bubble scatter plot using object oriencted interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="lightskyblue", s=[i * 5 for i in y_axis], alpha=0.2, edgecolors = "k", linewidths=2,label = "Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
#Create pie chart
plt.pie(y_axis, labels=x_axis)
```

```python
# create pie chart with style
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
#create pie chart with style and set the colors for each month

# Assign 12 colors, one for each month.
colors = ["royalblue", "deeppink", "green", "cyan", "yellow", "pink", "red", "olive", "gold", "orange", "peru", "crimson"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# pie chart using object oriented interface
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
#plot error bar
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
#plot error bar with caps
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#Plot error bars with object oriented 
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#create bar chart with error bars
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#create bar charts with error bars using object oriented interface
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
#change ticks
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
#object oriented approach ^^
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
#create minor ticks

from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
